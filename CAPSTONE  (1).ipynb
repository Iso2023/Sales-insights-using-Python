```python
#Convert the 'Quantity Ordered' and 'Price Each' columns to numeric data types
All_data['Quantity Ordered'] = pd.to_numeric(All_data['Quantity Ordered'])
All_data['Price Each'] = pd.to_numeric(All_data['Price Each'])

# Convert the 'Order Date' column to datetime data type
All_data['Order Date'] = pd.to_datetime(All_data['Order Date'], format='%m/%d/%y %H:%M')

# Verify the updated data types
print(All_data.dtypes)
```

```python
All_data['Total Sales'] = All_data['Quantity Ordered'] * All_data['Price Each']
print(All_data)
```

```python
# Create a new column for the 'Month' extracted from the 'Order Date'
All_data['Month'] = All_data['Order Date'].dt.month
```

```python
# Split the 'purchase address' column into 'street', 'city', and 'state' columns
All_data[['Street', 'City', 'State']] = All_data['Purchase Address'].str.split(',', expand=True)

# Strip whitespaces from the new columns
All_data['Street'] = All_data['Street'].str.strip()
All_data['City'] = All_data['City'].str.strip()
All_data['State'] = All_data['State'].str.strip()

# Print the DataFrame with the extracted 'street', 'city', and 'state'
print(All_data[['Purchase Address', 'Street', 'City', 'State']])

# Split the 'state' column based on whitespace and keep only the first part (state name)
All_data['State'] = All_data['State'].str.split().str[0]

# Print the DataFrame after removing the postal codes from the 'state' column
print(All_data[['Purchase Address', 'Street', 'City', 'State']])

```

```python
# Update the 'City' column for 'Portland' in Oregon (OR)
condition_or = (All_data['City'] == 'Portland') & (All_data['State'] == 'OR')
All_data.loc[condition_or, 'City'] = 'Portland OR'

# Update the 'City' column for 'Portland' in Maine (ME)
condition_me = (All_data['City'] == 'Portland') & (All_data['State'] == 'ME')
All_data.loc[condition_me, 'City'] = 'Portland ME'

# Print the DataFrame to see the updated 'City' column
All_data['City'].unique()
```

```python
##BEST MONTH PER SALES PER CITY###
import matplotlib.pyplot as plt

# Create a loop to plot line graphs for each city's data
for city in All_data['City'].unique():
    # Filter data for the current city
    city_data = All_data[All_data['City'] == city]
    
    # Group the data by 'State' and 'Month', and sum the 'Total Sales' column
    monthly_sales = city_data.groupby(['State', 'Month'])['Total Sales'].sum().reset_index()
    

    # Sort the DataFrame by the 'Month' column
    monthly_sales.sort_values('Month', inplace=True)
    
      
    # Plot a line graph showing the revenue in the current city by months
    plt.figure(figsize=(15, 6))
    for state, data in monthly_sales.groupby('State'):
        plt.plot(data['Month'], data['Total Sales'], marker="o", label=state,color='red')
    plt.xlabel('Month')
    plt.ylabel('Revenue')
    plt.title(f'Sales in {city} by Months')
    plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    plt.legend()
    plt.show()

```

```python
##BEST HOUR TO DISPLAY ADEVERTISEMENTS##

# Extract the hour from 'Order Date' and create a new column 'Order Hour'
All_data['Order Hour'] = All_data['Order Date'].dt.hour

# Group the data by 'Order Hour' and calculate the total sales for each hour
hourly_sales = All_data.groupby('Order Hour')['Total Sales'].sum()

# Plot a  chart to visualize the number of orders per hour
plt.figure(figsize=(12, 6))
plt.plot(hourly_sales.index, hourly_sales.values, marker=0, color='red')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales')
plt.title('Top sales by Hour')
plt.xticks(hourly_sales.index)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()
```

```python
# Calculate the highest sales month and its total sales for all cities
highest_sales_data = []

for city in All_data['City'].unique():
    # Filter data for the current city
    city_data = All_data[All_data['City'] == city]
    
    # Group the data by 'Month' and calculate the sum of 'Total Sales'
    monthly_sales = city_data.groupby('Month')['Total Sales'].sum().reset_index()
    
    # Find the highest sales month and its total sales
    max_sales_month = monthly_sales.loc[monthly_sales['Total Sales'].idxmax()]
    
    # Append the results to the 'highest_sales_data' list
    highest_sales_data.append({
        'City': city,
        'Highest Sales Month': max_sales_month['Month'],
        'Total Sales': max_sales_month['Total Sales']
    })

# Create a new DataFrame with the results
highest_sales_df = pd.DataFrame(highest_sales_data)

# Display the new DataFrame as a table
print(highest_sales_df)
```

```python
from prophet import Prophet
```

```python
##FORECAST FOR THREE MONTHS##
def forecast_revenue(data):
    model = Prophet(seasonality_mode='multiplicative', yearly_seasonality=False)
    model.fit(data)
    future = model.make_future_dataframe(periods=3 * 30)
    forecast = model.predict(future)
    return forecast

# Drop the time, keep only date
All_data['Order Date'] = pd.to_datetime(All_data['Order Date'])
All_data['Order Date'] = All_data['Order Date'].dt.date

# Create a list to store the forecasted sales for each state
predicted_revenue_list = []

# Group the data by 'State' and forecast sales for each state
for state, state_data in All_data.groupby('State'):
    state_data_copy = state_data[['Order Date', 'Total Sales']].copy()  # Create a copy of the slice
    state_data_copy.rename(columns={'Order Date': 'ds', 'Total Sales': 'y'}, inplace=True)
    forecast = forecast_revenue(state_data_copy)
    predicted_revenue_list.append((state, forecast[['ds', 'yhat']].tail(3)))

    # Plot the forecasted sales for each state
    plt.figure(figsize=(10, 6))
    plt.plot(forecast['ds'], forecast['yhat'], label=f'Forecast for - {state}')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.title(f'Forcasted sales for the next 3 Months - {state}')
    plt.legend()
    plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    plt.show()
```
